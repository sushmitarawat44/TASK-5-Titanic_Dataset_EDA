```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Pretty plots
sns.set(style="whitegrid")
```

```python
df = pd.read_csv("titanic (2)/train.csv")  
df.head()
```

```python
#dataset overview
df.info()
df.describe()
```

```python
#Categorial Distribution
df['Survived'].value_counts()
df['Pclass'].value_counts()
df['Sex'].value_counts()
df['Embarked'].value_counts()
```

```python
#pairplot
sns.pairplot(df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Survived')
```

```python
#heat map 
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
```

```python
#histogram of age
plt.figure(figsize=(8, 5))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title("Age Distribution")
```

```python
#survival by Gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
```

```python
#fare by class
sns.boxplot(x='Pclass', y='Fare', data=df)
plt.title("Fare Distribution by Passenger Class")
```

```python
#embarked port
sns.countplot(x='Embarked', hue='Survived', data=df)
plt.title("Survival by Embarkation Port")
```

```python
#cabin count values (Unknown included)
df['Cabin'].value_counts().head(10)
```

Observations 
Most passengers did not survive (~61.6%).

- Females had significantly higher survival rate.

- Higher class (Pclass 1) passengers had better chances of survival.

- Embarkation point C had a slightly better survival rate.

- Median Age is 28 (as I filled manually).

- Many Cabin values are "Missing", making it hard to use directly.So i filled the missing values with "Unknown" to make the data more useful
